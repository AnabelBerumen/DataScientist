```python
import pandas as pd
import numpy as np 
import io
```

```python
 df = pd.read_csv('/content/cancer.csv')
 df.head()
```

```python
X = df.iloc[:, 2:-1]
Y = df.iloc[:,1]
```

```python
X.head()
```

```python
X['Unnamed: 32'].isnull().sum()
```

```python
Y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=4 )
modelo = DecisionTreeClassifier()
modelo.fit(X_train, y_train)
resultado = modelo.score(X_test, y_test)
print(resultado)
```

```python
from sklearn.model_selection import KFold
from sklearn import tree
modelo = DecisionTreeClassifier()
kfold_validacion = KFold(10)
```

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(modelo, X, Y, cv = kfold_validacion)
print(resultados)
resultados.mean()
```

```python
from io import StringIO
from IPython.display import Image, display
import pydotplus
from sklearn import tree

out = StringIO()
# modelo es el entrenado
tree.export_graphviz(modelo, out_file = out)

# creamos las ramas que tenemos
graph = pydotplus.graph_from_dot_data(out.getvalue())
# creamos el archivo png
graph.write_png('cancer.png')
```
