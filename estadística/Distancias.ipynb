```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv('/content/movies.csv', sep=';')
data
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1)
dd2 = distance_matrix(data[movies], data[movies], p=2)
dd10 = distance_matrix(data[movies], data[movies], p=10)
```

```python
def dm_to_df(dd, col_name):
  import pandas as pd
  return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
# manhathan
dm_to_df(dd1, data['user_id'])
```

```python
# euclidea
dm_to_df(dd2, data['user_id'])
```

```python
# 10 minkowski
dm_to_df(dd10, data['user_id'])
```

### Clustering Jerárquico

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], 'ward')
Z
```

```python
plt.figure(figsize=(25, 10))
plt.title('Dendograma jerarquico para el clustering')
plt.xlabel('ID de los usuarios')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90., leaf_font_size=10)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.title('Dendograma jerarquico para el clustering')
plt.xlabel('ID de los usuarios')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90., leaf_font_size=10, orientation='right')
plt.show()
```

```python
Z = linkage(data[movies], 'average')
plt.figure(figsize=(25, 10))
plt.title('Dendograma jerarquico para el clustering')
plt.xlabel('ID de los usuarios')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=90., leaf_font_size=10)
plt.show()
```
